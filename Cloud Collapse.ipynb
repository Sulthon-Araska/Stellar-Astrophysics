```python
# @title **Impor Libraries**

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as sc
from scipy.integrate import solve_ivp
from IPython.display import display

# Constants:
π  = sc.pi
G  = sc.G
pc = sc.parsec

def Graph():
  plt.tight_layout()
  plt.legend()
  plt.grid()
```

```python
# Timelapse

# Types of Molecular Clouds
ρ = [5e-14, 1e-13, 5e-13]
R = [10*pc, 50*pc, 1*pc]

# Collapse rate (ordinary differential equation)
def ODE(i,r):
  t = -1/np.sqrt((8*π*G*ρ[i]*(R[i]**2)/3)*(r - 1))
  return t

# Solving differential equation
r1 = solve_ivp(ODE(1,r), [0,1], [1], method='DOP853')
```
