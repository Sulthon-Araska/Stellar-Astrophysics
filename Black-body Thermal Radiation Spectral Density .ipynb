```python
# @title **Frameworks**

# Import libraries
import numpy as np
import scipy.constants as sc
import matplotlib.pyplot as plt

# Physical constants
h = sc.h       # Planck constant
c = sc.c       # Lightspeed constant
k = sc.k       # Boltzmann constant
λ = np.linspace(0, 3e-6, 1000)

# Rayleigh-Jeans Law
def R1(T,v):    #... function of frequency
  B = (2*(v**2)*k*T)/(c**2)
  return B

def R2(T,λ):    #... function of wavelength
  B = (2*c*k*T)/(λ**4)
  return B

# Wien Approximation
def W1(T,v):    #... function of frequency
  B = (2*h*(v**3))*np.exp(-(h*v)/(k*T))/(c**2)
  return B

def W2(T,λ):    #... function of wavelength
  B = (2*h*(c**2))*np.exp(-(h*c)/(λ*k*T))/(λ**5)
  return B

# Planck's Law
def P1(T,v):    #... function of frequency
  B = (2*h*(v**3))/((c**2)*np.expm1((h*v)/(k*T)))
  return B

def P2(T,λ):    #... function of wavelength
  B = (2*h*(c**2))/((λ**5)*np.expm1((h*c)/(λ*k*T)))
  return B
```

## **A. *Spectral Density as a functions of frequency (υ)***

```python
# @title **Temperature = 6000 K**
v = np.linspace(0, 1.5e15, 1000)
T = 6000

# Functions:
Rayleigh_Jeans = R1(T, v)
Wien           = W1(T, v)
Planck         = P1(T, v)

# Line plot
plt.plot(v, Rayleigh_Jeans, label='Rayleigh-Jeans Law', color='r', linestyle='dashed')
plt.plot(v, Planck,         label='Planck Law',         color='g')
plt.plot(v, Wien,           label='Wien Approximation', color='b', linestyle='dotted')

# Miscellaneous
plt.title("Spectral Density as a function of Frequency at 6,000 K")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Spectral Radiance (W.sr-1.m-2.Hz-1)')
plt.ylim(0, 0.6e-7)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## **3D Graph of Spectral Power Distribution: Planck's Law of Black Body Thermal Radiation**

```python
# @title *Disclaimer:* Code made by **Bing AI**

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Constants
h = 6.626e-34  # Planck's constant
c = 3.0e+8     # Speed of light
k = 1.38e-23   # Boltzmann constant

def planck(wav, T):
    a = 2.0*h*c**2
    b = h*c/(wav*k*T)
    intensity = a/ ((wav**5) * (np.exp(b) - 1.0))
    return intensity

# Generate wavelength and temperature arrays
wav = np.arange(1e-9, 3e-6, 1e-9)
T = np.arange(3000, 8000, 100)

# Create a meshgrid for wavelength and temperature arrays
WAV, T = np.meshgrid(wav, T)

# Calculate intensity using Planck's law
I = planck(WAV, T)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(WAV, T, I, rstride=1, cstride=1, cmap=cm.coolwarm)
ax.set_xlabel('Wavelength (m)')
ax.set_ylabel('Temperature (K)')
ax.set_zlabel('Intensity')

plt.show()
```
